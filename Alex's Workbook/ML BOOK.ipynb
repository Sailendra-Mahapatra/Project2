```python
engine = create_engine('sqlite:///../data/china_trade.sqlite', echo=False)
```

```python
Base = automap_base()
Base.prepare(engine, reflect=True)
session = Session(engine)
```

```python
IndImports = Base.classes.hs2import
IndExports = Base.classes.hs2export
YRImports = Base.classes.yrhs2import
YRExports = Base.classes.yrhs2export
Imports = Base.classes.imports
Exports = Base.classes.export
```

```python
stmt = session.query(Imports).statement
df = pd.read_sql_query(stmt, session.bind)
```

```python
df['Period'] = pd.to_datetime(df['Period'])

df.Period = pd.to_numeric(df.Period, errors='coerce')
```

```python
y = df.YTDValue
```

```python
features = ['MoValue', 'Period', 'HSC']
X = df[features]
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.2, 
                                                    random_state=123
                                                    )
```

```python
imports_model = DecisionTreeRegressor(random_state=1)
```

```python
imports_model.fit(train_X, train_y)
```

```python
val_predictions = imports_model.predict(val_X)
```

```python
val_mae = mean_absolute_error(val_predictions, val_y)
```

```python
imports_model = DecisionTreeRegressor(max_leaf_nodes=100, random_state=1)
```

```python
imports_model.fit(train_X, train_y)
```

```python
print("Validation MAE for best value of max_leaf_nodes: {:,.0f}".format(val_mae))
```
